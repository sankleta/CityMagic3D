```python
resolution = (4864, 3648)
camera = Camera(intrinsic_matrix, *resolution)
```

```python
wmsc_pc_path = "/media/papjuli/Data/data/3DV_project/STPLS3D/point_clouds/RealWorldData/WMSC_points.ply"
pc_o3d = o3d.io.read_point_cloud(wmsc_pc_path)
```

```python
# img_name = "DJI_0029 (3).JPG"
# img_name = "DJI_0014 (3).JPG"
img_name = "DJI_0341.JPG"
img_pose = poses_for_images[img_name]
```

```python
camera_lines = o3d.geometry.LineSet.create_camera_visualization(
    view_width_px=resolution[0], view_height_px=resolution[1], 
    intrinsic=camera.intrinsic_matrix.copy(), 
    extrinsic=np.linalg.inv(img_pose),
    scale=20)
```

```python
for mask in mask_indices_ls:
    np.asarray(pc_o3d.colors)[mask,:] = (np.random.random(3) + np.asarray(pc_o3d.colors)[mask,:]) / 2
```

```python
o3d.visualization.draw_geometries([pc_o3d, camera_lines],
                                  zoom=0.34,
                                  front=[1, 0, 0.5],
                                  lookat=[0, 0, 0],
                                  up=[-0.5, 0, 1])
```
